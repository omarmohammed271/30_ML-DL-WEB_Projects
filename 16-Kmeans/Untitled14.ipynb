```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
# read data 
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
# elbow method
results = []
for i in range(1,11):
    model = KMeans(n_clusters=i)
    model.fit(x)
    results.append(model.inertia_)
    
results
```

```python
plt.plot(range(1,11),results)
plt.title('Elbow Method')
plt.xlabel('Number of Clsuters')
plt.ylabel('Result')
```

```python
model = KMeans(n_clusters=5)
model.fit(x)
```

```python
y_labels = model.predict(x)
```

```python
# test predict
model.predict([[15,70]])
```

```python
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=y_labels,cmap='rainbow')
```
