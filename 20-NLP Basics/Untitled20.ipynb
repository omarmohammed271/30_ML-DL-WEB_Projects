### stemming 

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

words = ['going','like','likes','went','better']

for i in words:
    print(f"{i} - {ps.stem(i)}")
```

---

## limmatization

```python
from nltk.stem import WordNetLemmatizer

lm = WordNetLemmatizer()

words = ['going','like','likes','went','better']

for i in words:
    print(f"{i} - {lm.lemmatize(i)}")

```

```python
from nltk.corpus import wordnet

lm.lemmatize('going')
```

```python
lm.lemmatize('going',wordnet.VERB)
```

```python
lm.lemmatize('better')
```

```python
lm.lemmatize('better',wordnet.ADJ)
```

### Stop Words

```python
from nltk.corpus import stopwords

#nltk.download('stopwords')

stopwords.words('english')
```

```python
stopwords.words('arabic')
```

```python
# tokanize remove stopword
data = "my name is mahmoud , i love coding , i am a python developer"
mystopwords = stopwords.words('english')

result = []

for x in nltk.word_tokenize(data):
    if x not in mystopwords :
        result.append(x)
        
        
result
```

```python
result = [x for x in nltk.word_tokenize(data) if x not in mystopwords]
result
```
