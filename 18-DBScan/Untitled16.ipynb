```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
```

```python
# read data 
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],cmap='rainbow')
```

```python
## Scalling 
from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(x)
```

```python
model = DBSCAN(min_samples=5 , eps=.4)
y_lables = model.fit_predict(x_scaled)
```

```python
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=y_lables,cmap='rainbow')
```
